```python
# Primeiro modelo de aprendizado supervisionado
# Regressão Linear
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Coleta de dados:
df = pd.read_csv("alunos.csv")

# Variaveis independentes e dependentes
x = df[["horas_estudo"]]
y = df["nota"]

#Treinamento do modelo de Regressão Linear
modelo = LinearRegression()
modelo.fit(x, y)

# Coeficientes: inclinação e intercepto
print(f'Inclinação: {modelo.coef_[0]}')
print(f'Intercepto: {modelo.intercept_}')

# Previsão
y_previsto = modelo.predict(x)

# Gráfico
y_pred = modelo.predict(x)

plt.scatter(x, y, label="Dados reais: Horas de estudo")
plt.plot(x, y_pred, color="red", label="Regressão Linear")
plt.xlabel("Horas de Estudo")
plt.ylabel("Nota")
plt.legend()
plt.show()

horas = [[6]]
nota_prevista = modelo.predict(horas)
print(f'Nota prevista: {nota_prevista[0]} para {horas} horas de estudo.')
```

```python
# Primeiro modelo de aprendizado supervisionado
# Regressão Linear
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Coleta de dados:
df = pd.read_csv("alunos.csv")

# Variaveis independentes e dependentes
x = df[["faltas"]]
y = df["nota"]

#Treinamento do modelo de Regressão Linear
modelo = LinearRegression()
modelo.fit(x, y)

# Coeficientes: inclinação e intercepto
print(f'Inclinação: {modelo.coef_[0]}')
print(f'Intercepto: {modelo.intercept_}')

# Previsão
y_previsto = modelo.predict(x)

# Gráfico
y_pred = modelo.predict(x)

plt.scatter(x, y, label="Dados reais: Faltas")
plt.plot(x, y_pred, color="red", label="Regressão Linear")
plt.xlabel("Faltas")
plt.ylabel("Nota")
plt.legend()
plt.show()

faltas = [[6]]
nota_prevista = modelo.predict(faltas)
print(f'Nota prevista: {nota_prevista[0]} para quantidade de {faltas} faltas.')
```
